```python
import pandas_datareader.data as web
```

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
import datetime as dt
```

```python
df = web.DataReader('GE', 'yahoo', start='2021-09-09', end='2022-09-09')
```

```python
close = df['Close']
```

```python
all_weekdays = pd.date_range(start='2021-09-09', end='2022-09-09', freq='B')
```

```python
close = close.reindex(all_weekdays)
```

```python
close = close.fillna(method='ffill')
```

```python
data = close.loc[:]
```

```python
short_rolling_msft = data.rolling(window=20).mean()
long_rolling_msft = data.rolling(window=10).mean()
```

```python
fig, ax = plt.subplots(figsize=(16,9))

ax.plot(data.index, data, label='YHOO')
ax.plot(short_rolling_msft.index, short_rolling_msft, label='20 days rolling')
ax.plot(long_rolling_msft.index, long_rolling_msft, label='10 days rolling')

ax.set_xlabel('Date')
ax.set_ylabel('Adjusted closing price ($)')
ax.legend()
ax.grid()
```
